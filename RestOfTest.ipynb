```python
customers = pd.read_csv("customers.csv")
items = pd.read_csv("items.csv")
orderItem = pd.read_csv("order_item.csv")
orders = pd.read_csv("orders.csv")
#graphs 3 to 7

#3
sns.displot(x = customers['age'], color = 'maroon')
plt.xticks(rotation=45, fontsize=10)
plt.title("Age Distribution")
plt.show()

```

```python
#4
customers['joining_date'] = pd.to_datetime(customers['joining_date'], format = "%d/%m/%Y")
yearlyCustomers = customers.resample(rule = 'YS', on = 'joining_date').size()
sns.barplot(x = yearlyCustomers.index, y=yearlyCustomers.values, color = 'skyblue')
plt.xticks(rotation=45, fontsize=10)
plt.title("Yearly New Customers")
plt.show()

```

```python
#5
orders['order_date'] = pd.to_datetime(orders['order_date'])
monthlySales = orders.resample(rule = 'MS', on = 'order_date').size()
sns.barplot(x = monthlySales.index, y=monthlySales.values, color = 'gold')
plt.xticks(rotation=45, fontsize=10)
plt.title("Monthly Sales")
plt.show()

```

```python
#6
orderCustomerMerge = orders.merge(customers, left_on = 'customer_id', right_on = 'id')
orderCustomer = orderCustomerMerge.groupby('age').size()
sns.displot(x = orderCustomerMerge['age'], color = 'gold')
plt.xticks(rotation=45, fontsize=10)
plt.title("Sales By Age")
plt.show()

#The sales number increase towared 45, peaks there and then goes down, which mean that around the ages 45 and 55 are the highest sales.

orderCustomerMerge
```

```python
#7 bonus
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter



MegaMerge = orderCustomerMerge.merge(orderItem, on = 'order_id').merge(items, left_on = 'item_id', right_on = 'id')
MegaMerge['total_price'] = MegaMerge['quantity'] * MegaMerge['item_price']
finalMerge = MegaMerge.groupby('nationallity')['total_price'].sum()
plt.ticklabel_format(style='plain') #found this on stack overflow, it shows the full number 
sns.barplot(x = finalMerge.index, y = finalMerge.values, color = 'silver')
plt.xticks(rotation=45, fontsize=10)
plt.title("Sales By Nationallity")
plt.show()
finalMerge
```
